```python
from pyspark.sql import functions as sf
```

```python
energy_table = spark.read.table('our_world_in_data.bronze_layer.energy')
```

```python
# list columns types
biofuel_columns = []
coal_columns = []
electricity_columns = []
fossil_columns = []
gas_columns = []
hydro_columns = []
low_carbon_columns = []
nuclear_columns = []
oil_columns = []
other_renewable_columns = []
renewables_columns = []
solar_columns = []
wind_columns = []
energy_columns = []
other_columns = []

for columns in energy_table.columns:
    if(('biofuel' in columns) and ('_biofuel' not in columns)):
        biofuel_columns.append(columns)
    elif(('coal' in columns) and ('_coal' not in columns)):
        coal_columns.append(columns)
    elif(('electricity' in columns) and ('_electricity' not in columns)):
        electricity_columns.append(columns)
    elif(('fossil' in columns) and ('_fossil' not in columns)):
        fossil_columns.append(columns)
    elif(('gas' in columns) and ('_gas' not in columns)):
        gas_columns.append(columns)
    elif(('hydro' in columns) and ('_hydro' not in columns)):
        hydro_columns.append(columns)
    elif(('low_carbon' in columns) and ('_low_carbon' not in columns)):
        low_carbon_columns.append(columns)
    elif(('nuclear' in columns) and ('_nuclear' not in columns)):
        nuclear_columns.append(columns)
    elif(('oil' in columns) and ('_oil' not in columns)):
        oil_columns.append(columns)
    elif(('other_renewable' in columns) and ('_other_renewable' not in columns)):
        other_renewable_columns.append(columns)
    elif(('renewables' in columns) and ('_renewables' not in columns)):
        renewables_columns.append(columns)
    elif(('solar' in columns) and ('_solar' not in columns)):
        solar_columns.append(columns)
    elif(('wind' in columns) and ('_wind' not in columns)):
        wind_columns.append(columns)
    elif(('energy' in columns) and ('_energy' not in columns)):
        energy_columns.append(columns)
    else:
        other_columns.append(columns)
```

```python
%sql
DROP TABLE IF EXISTS our_world_in_data.silver_layer.country;

CREATE TABLE IF NOT EXISTS our_world_in_data.silver_layer.country(
  Country_id BIGINT GENERATED ALWAYS AS IDENTITY,
  CountryName VARCHAR(300),
  CountryIsoCode VARCHAR(3)
);
```

```python
%sql
INSERT INTO our_world_in_data.silver_layer.country(
  CountryName,
  CountryIsoCode
)
SELECT
  country,
  iso_code
FROM
  our_world_in_data.bronze_layer.energy
WHERE country not like '%(%' AND iso_code IS NOT NULL
GROUP BY
  country,
  iso_code
ORDER BY  
  country;
```
